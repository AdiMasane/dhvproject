```python
df = df.drop(['pm2_5'], axis=1)
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_') #remove extra spaces
```

```python
pollutants = ['so2', 'no2', 'rspm', 'spm']  # change based on your exact columns

for pollutant in pollutants:
    df[pollutant] = df[pollutant].fillna(df[pollutant].mean())
```

```python
df['stn_code'] = df['stn_code'].fillna('Unknown')
df['agency'] = df['agency'].fillna('Unknown')
df['type'] = df['type'].fillna('Unknown')
df['location_monitoring_station'] = df['location_monitoring_station'].fillna('Unknown')
```

```python
df['location'] = df['location'].fillna(df['location'].mode()[0])
```

```python
# Fix 'sampling_date' and 'date'
df['sampling_date'] = pd.to_datetime(df['sampling_date'], errors='coerce')
df['sampling_date'] = df['sampling_date'].fillna(pd.to_datetime('2000-01-01'))

df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['date'] = df['date'].fillna(df['date'].min())


```

```python
# Extract year and month
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
```

```python
# Remove outliers using Z-score method
#Z= σ/(X−μ)
from scipy.stats import zscore
df_z = df[pollutants].apply(zscore)
df = df[(df_z.abs() < 3).all(axis=1)]
```

```python
grouped = df.groupby(['state', 'location'])[pollutants].mean().reset_index()
grouped.head()
```

```python
monthly = df.groupby(['state', 'location', 'year', 'month'])[pollutants].mean().reset_index()
monthly.head()
```

```python
grouped.to_csv('cleaned_air_quality.csv', index=False)
monthly.to_csv('monthly_air_quality.csv', index=False)
```

```python
# Q1: Which states have the highest average SO2 levels?
print("Q1: Which states have the highest average SO2 levels?")
print("Answer: These are the top states based on SO2 average:")
state_so2 = df.groupby('state')['so2'].mean().sort_values(ascending=False).head(10)
print(state_so2)
state_so2.plot(kind='bar', figsize=(10, 5), color='orange')
plt.title('Top 10 States with Highest SO2 Levels')
plt.ylabel('Average SO2')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Q2: Which cities have the highest NO2 levels?
print("Q2: Which cities have the highest NO2 levels?")
print("Answer: These cities show the highest NO2 average levels:")
city_no2 = df.groupby('location')['no2'].mean().sort_values(ascending=False).head(10)
print(city_no2)
city_no2.plot(kind='bar', figsize=(10, 5), color='purple')
plt.title('Top 10 Cities with Highest NO2 Levels')
plt.ylabel('Average NO2')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Q3: Distribution of SPM levels across India
print("Q3: How are SPM levels distributed across India?")
print("Answer: The histogram below shows the distribution of SPM levels across all data.")
df['spm'].plot(kind='hist', bins=30, figsize=(10, 5), color='green')
plt.title('Distribution of SPM Levels')
plt.xlabel('SPM (µg/m³)')
plt.tight_layout()
plt.show()
```

```python
# Q4: What are the average NO2 levels per state?
print("Q4: What are the average NO2 levels per state?")
print("Answer: NO2 state-wise averages are shown below.")
state_no2 = df.groupby('state')['no2'].mean().sort_values(ascending=False)
print(state_no2.head(10))
state_no2.plot(kind='bar', figsize=(12, 5), color='teal')
plt.title('Average NO2 Levels by State')
plt.ylabel('NO2')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Q5: Which monitoring types (urban/rural/etc.) have more pollution?
print("Q5: Pollution by monitoring type")
print("Answer: Average SO2 and NO2 levels by type are shown.")
type_avg = df.groupby('type')[['so2', 'no2']].mean()
print(type_avg)
type_avg.plot(kind='bar', figsize=(10, 5))
plt.title('Average SO2 and NO2 by Monitoring Type')
plt.ylabel('µg/m³')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Question
print("Q6: Which agencies are monitoring the cleanest areas?")

# Answer: Identify the agencies monitoring areas with the lowest pollutant averages (PM2.5, SO2, NO2)
agency_pollution = df.groupby('agency')[[ 'so2', 'no2']].mean()
cleanest_agencies = agency_pollution.mean(axis=1).sort_values().head(10)

# Graph
cleanest_agencies.plot(kind='bar', figsize=(10, 6), color='green')
plt.title('Agencies Monitoring Cleanest Areas')
plt.xlabel('Agency')
plt.ylabel('Average Pollution Level (µg/m³)')
plt.show()
```

```python
# Q7: Top 10 most polluted locations based on NO2
print("Q7: Top 10 polluted locations based on NO2")
print("Answer: These locations have highest NO2 on average.")
location_no2 = df.groupby('location')['no2'].mean().sort_values(ascending=False).head(10)
print(location_no2)
location_no2.plot(kind='bar', figsize=(10, 5), color='red')
plt.title('Top 10 Polluted Locations (NO2)')
plt.ylabel('Average NO2')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Q8: State-wise NO2 vs SO2 correlation
print("Q8: State-wise correlation between NO2 and SO2")
print("Answer: Scatter plot shows relation between NO2 and SO2 by state average.")
cor_df = df.groupby('state')[['no2', 'so2']].mean()
plt.figure(figsize=(8, 5))
plt.scatter(cor_df['no2'], cor_df['so2'], color='orange')
plt.xlabel('Average NO2')
plt.ylabel('Average SO2')
plt.title('Correlation between NO2 and SO2 by State')
plt.tight_layout()
plt.show()
```

```python
# Q9: Which states have highest average RSPM?
print("Q9: States with highest RSPM")
print("Answer: These states show high average RSPM levels:")
state_rspm = df.groupby('state')['rspm'].mean().sort_values(ascending=False).head(10)
print(state_rspm)
state_rspm.plot(kind='bar', figsize=(10, 5), color='gray')
plt.title('Top 10 States with Highest RSPM')
plt.ylabel('Average RSPM')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Q10: Outlier cities with high SPM
print("Q10: Outlier cities with high SPM")
print("Answer: The boxplot highlights outliers for SPM levels across cities.")
plt.figure(figsize=(10, 5))
df.boxplot(column='spm', by='location', showfliers=True, rot=90)
plt.title('SPM Outliers by Location')
plt.suptitle('')
plt.ylabel('SPM (µg/m³)')
plt.tight_layout()
plt.show()
```

```python
# Question
print("Q11: Identify outlier cities with very high SPM")

# Answer: Cities with extremely high SPM values can be visualized as outliers in this scatter plot

# Calculate average SPM by city
city_spm = df.groupby('location')['spm'].mean().sort_values(ascending=False)

# Graph
plt.figure(figsize=(12, 6))
plt.scatter(city_spm.index, city_spm.values, color='tomato')
plt.xticks(rotation=90)
plt.ylabel("Average SPM (µg/m³)")
plt.title("Cities with Highest Average SPM Levels")
plt.grid(True)
plt.tight_layout()
plt.show()

# Print top 10 outlier cities
print("\nTop 10 Cities with Highest Average SPM:")
print(city_spm.head(10))
```
